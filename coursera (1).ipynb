```python
import numpy as np 

import pandas as pd 
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json 
import requests 
from pandas.io.json import json_normalize 
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
import folium 

print('Libraries imported.')
```

```python
table = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
```

```python
df=table[0]
```

```python
dfd1=df[df.Borough!="Not assigned"]
```

```python
geo=pd.read_csv("http://cocl.us/Geospatial_data")
geo
```

```python
k=df["Postal Code"]
pc=list(k.values)
lat=[]
lon=[]
```

```python
for n in pc:
    lat.append(float(geo[geo["Postal Code"]==n]["Latitude"].values))
    lon.append(float(geo[geo["Postal Code"]==n]["Longitude"].values))
```

```python
df["Latitude"]=lat
df["Longitude"]=lon
```

```python
kclusters = 5
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(df[["Latitude","Longitude"]])
```

```python
df.insert(0, 'Cluster Labels', kmeans.labels_)
df
```

#### So above here we get the final version of our dataframe with the updated latitude and longitude

```python
# create map
map_clusters = folium.Map(location=[43.716316,-79.239476], zoom_start=10)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(df['Latitude'], df['Longitude'], df['Neighborhood'], df['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
